```python
from sklearn.preprocessing import MinMaxScaler
from ucimlrepo import fetch_ucirepo
import pandas as pd

# Carregar base de dados
df1 = fetch_ucirepo(id=519)

#df1.data
```

```python
import numpy as np

# Definição de 'X' e 'y' no Dataset
X = df1.data.features
y = df1.data.targets

# Verifica a quantidade de classes tem 'y'
classes = df1.data.targets.value_counts().shape[0]

print("Total de classes: " + str(classes))
print('*'*50 + 'Target' + '*'*50)
display(y)
print('*'*50 + 'Features' + '*'*50)
display(X)
```

```python
import numpy as np
import pandas as pd

# Verificação de NaN nas Colunas
total_nan = X.isna().sum()

# Para verificar a porcentagem de Nan em cada coluna
#porcentagem_de_na = X.isna().sum() / len(X) * 100

#print("Porcentágens de NaN nas Colunas: {:.2f}%".format(porcentagem_de_na))
print('Total de NaN nas Colunas:')
print(total_nan)
```

```python
from sklearn import preprocessing

# Normalização das Features
dfx = X.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(dfx)
X = pd.DataFrame(x_scaled)

display(X)
print(y)
```

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Transforma para Array NumPy
X = np.array(X)
y = np.array(y)
y = y.ravel()

# Total de clusters
folds = 10

kf = StratifiedKFold(n_splits = folds)

X_train = []
y_train = []

X_test = []
y_test = []

for train_index, test_index in kf.split(X,y):
  X_train.append(X[train_index])
  X_test.append(X[test_index])

  y_train.append(y[train_index])
  y_test.append(y[test_index])

#X_train
#y_train
y_test
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

results1 = []
results2 = []

erros1 = []
erros2 = []

for i in range(folds):
  model = DecisionTreeClassifier(criterion="entropy")
  treino = model.fit(X_train[i], y_train[i])

  result = treino.predict(X_test[i])

  acc = metrics.accuracy_score(result, y_test[i])
  err = mean_squared_error(y_test[i], result)

  results1.append(acc)
  erros1.append(err)

for i in range(folds):
  model = DecisionTreeClassifier(criterion="gini")
  treino = model.fit(X_train[i], y_train[i])

  result = treino.predict(X_test[i])

  acc = metrics.accuracy_score(result, y_test[i])
  err = mean_squared_error(y_test[i], result)

  results2.append(acc)
  erros2.append(err)

show1 = round(np.mean(results1) * 100)
show2 = round(np.mean(results2) * 100)

erro1= np.mean(erros1)
erro2= np.mean(erros2)

print("Acurácia Árvore com Entropy: {:.2f}%".format(show1))
print(results1)
print("Erro Árvore com Entropy: {:.2f}".format(erro1))
print(erros1)

print()
print("Acurácia Árvore com Gini: {:.2f}%".format(show2))
print(results2)
print("Erro Árvore com Gini: {:.2f}".format(erro2))
print(erros2)
```

```python
from typing import Mapping
from sklearn.cluster import KMeans
from collections import Counter

# Verifica a quantidade de 'clusters'
clusters = df1.data.targets.value_counts().shape[0]

# Aplit Dataset
results9 = []
erros9 = []
resulteste = []

for i in range(folds):
  model = KMeans(n_clusters = clusters)
  treino = model.fit(X_train[i])

  # Pegar os labels dos padrões de Treinamento
  labels = treino.labels_

  map_labels = []

  for j in range(clusters):
    map_labels.append([])

  new_y_train = y_train[i]

  for j in range(len(y_train[i])):
    for c in range(clusters):
        if labels[j] == c:
            map_labels[c].append(new_y_train[j])

  # Criar dicionário com os labells a serem mapeados
  mapping = {}

  for j in range(clusters):
    final = Counter(map_labels[j]) # contar a classe que mais aparece
    value = final.most_common(1)[0][0] # retorna a classe com maior frequência
    mapping[j] = value

  result = treino.predict(X_test[i])
  result = [mapping[i] for i in result]

  acc = metrics.accuracy_score(result, y_test[i])
  err = mean_squared_error(y_test[i], result)

  results9.append(acc)
  erros9.append(err)
  resulteste.append(result)


show9 = round(np.mean(results9) * 100)
erro9 = np.mean(erros9)

#print(clusters)
print("Acurácia 2-Means: {:.2f}%".format(show9))
print(results9)

print()
print(list(resulteste))
print(list(y_test))

print()
print("Erro 2-Means: {:.2f}".format(erro9))
print(erros9)
```
